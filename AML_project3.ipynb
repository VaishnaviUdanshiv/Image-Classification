Loading Required Libraries

```python
import os
import cv2
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

Pre-processing phase

```python
train_dir = 'D://MachineLearning//Project3_files//Project3_files//data//train'  # train dataset directory
test_dir = 'D://MachineLearning//Project3_files//Project3_files//data//test'  # test dataset directory
image_size = (350, 350)  # image size
```

```python
#Loading training data
train_images = []
train_labels = []

for image_name in os.listdir(train_dir):
    image_path = os.path.join(train_dir, image_name)
    img = cv2.imread(image_path)
    img = cv2.resize(img, image_size)
    img = img.astype('float32') / 255.0  # Normalization
    train_images.append(img)
    label = image_name.split('.')[0]  # Extracting the label 
    train_labels.append(label)
```

```python
#Loading testing data
test_images = []
test_labels = []

for image_name in os.listdir(test_dir):
    image_path = os.path.join(test_dir, image_name)
    img = cv2.imread(image_path)
    img = cv2.resize(img, image_size)
    img = img.astype('float32') / 255.0  # Normalization
    test_images.append(img)
    label = image_name.split('.')[0]  # Extracting the label 
    if label == 'cat':
        test_labels.append(0)  # Assigning 0 for cat label
    else:
        test_labels.append(1)  # Assigning 1 for dog label
```

```python
# Convert the train and test images to numpy arrays
train_images = np.array(train_images)   #Train images to numpy array
test_images = np.array(test_images)    #test images to numpy array

label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_labels)  #label encoding on training set

train_images, train_labels = shuffle(train_images, train_labels)  #shuffling training set
```

Training Phase

```python
model = Sequential()    # 2. Creating a the CNN Model
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(image_size[0], image_size[1], 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  

model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])   # Compiling the model

model.fit(train_images, train_labels, epochs=10, batch_size=32)  # Training the CNN Model
```

Evaluation phase

```python
test_predictions = model.predict(test_images)   #predicting on testing set
test_predictions = np.round(test_predictions).astype(int)

for i in range(len(test_predictions)):
    if test_predictions[i] == 0:
        predicted_label = 'cat'
    else:
        predicted_label = 'dog'
    print("True Label:", test_labels[i])
    print("Predicted Label:", predicted_label)
    print()

accuracy = np.mean(test_predictions == test_labels)  # Calculating accuracy score
print("Test Accuracy:", accuracy)
```

Hyperparamater Optimization

```python
def create_model(optimizer='adam', activation='relu'):   #creating a function for model
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation=activation, input_shape=(image_size[0], image_size[1], 3)))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(64, (3, 3), activation=activation))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(128, (3, 3), activation=activation))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(128, activation=activation))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32)   # Creating the KerasClassifier wrapper for scikit-learn compatibility

param_grid = {    
    'optimizer': ['adam', 'rmsprop'],    # Defining the hyperparameters to tune
    'activation': ['relu', 'sigmoid']
}

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)   # Performing grid search with cross-validation
grid_result = grid.fit(train_images, train_labels)

print("Best Hyperparameters: ", grid_result.best_params_)   # Printing the best hyperparameters 
print("Best Score: ", grid_result.best_score_)     #printing the results
 
```
